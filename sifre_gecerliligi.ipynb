```python
#ŞİFRE GEÇERLİ Mİ?

kontrol=0
ozel_isaret="!#+%&?_"

sifre=input("Lütfen bir şifre giriniz: ")

while True:
    #uzunluk kontrolü
    if len(sifre) >= 8 and len(sifre) <=10:

        #küçük harf kontrolü
        for i in sifre:
            if i >="a" and i <= "z":
                kontrol+=1
                break

        if kontrol==1: # küçük harf kontrolünden geçtiyse
            #büyük harf kontrolü
            for i in sifre:
                if i >="A" and i <= "Z":
                    kontrol+=1
                    break

        if kontrol==2: # büyük harf kontrolünden geçtiyse
            for i in sifre:
                if i in ozel_isaret:
                    kontrol+=1
                    break


        if kontrol==3: # özel karakter kontrolünden geçtiyse
            for i in sifre:  # sayı kontrolü
                if i.isnumeric():
                    kontrol+=1
                    break

        if kontrol==4:
            print("Şifre geçerli")
            break
        else:
            print("Şifre geçersiz")
            break


    else:
        print("Şİfreniz 8 ile 10 karakter arasında olmalıdır")
        break

```
